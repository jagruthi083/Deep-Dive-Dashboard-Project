# Deep Dive Analysis Notebook
This notebook performs KPI analysis and cohort analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/data.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'])
df.isnull().sum()
```

## KPI Calculations

```python
conversion_rate = len(df[df['orders'] > 1]) / len(df) * 100
churn_rate = df['churned'].mean() * 100
aov = df['revenue'].sum() / df['orders'].sum()

print('Conversion Rate:', round(conversion_rate, 2), '%')
print('Churn Rate:', round(churn_rate, 2), '%')
print('Average Order Value:', round(aov, 2))
```

## Revenue Trend

```python
plt.figure()
df.groupby('signup_date')['revenue'].sum().plot()
plt.title('Revenue Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```

## Cohort Analysis

```python
df['signup_month'] = df['signup_date'].dt.to_period('M')
cohort = df.groupby('signup_month')['revenue'].sum()

cohort.plot(kind='bar')
plt.title('Cohort Revenue Analysis')
plt.xlabel('Signup Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

## Churn Distribution

```python
sns.countplot(x='churned', data=df)
plt.title('Churn Distribution')
plt.show()
```

## Orders vs Revenue

```python
plt.scatter(df['orders'], df['revenue'])
plt.title('Orders vs Revenue')
plt.xlabel('Orders')
plt.ylabel('Revenue')
plt.show()
```

## Key Insights
- Higher orders lead to higher revenue
- Churn rate is around 30%
- Revenue varies across months
- Some cohorts perform better
